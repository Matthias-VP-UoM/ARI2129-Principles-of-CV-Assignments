# Part 1: Point Processing

## Task 1 - Loading grayscale image

```python
# Reference to image
image_name = "player.png"

# Loading image using cv2.imread()
img = cv2.imread(image_name, cv2.IMREAD_GRAYSCALE)
```

## Task 2 - Implementing transformation functions

### a) Logarithmic Transform

```python
def apply_log_transform(r, a):
    img_log = (np.log(r+1))*255

    s = np.array(img_log * a ,dtype=np.uint8)

    return s
```

### b) Power Transform

```python
def apply_power_transform(r, gamma, c):
    s = np.array(c*(r/255)**gamma,dtype=np.uint8)
    #s = np.array(c*r**gamma,dtype=np.uint8)
    return s
```

### c) Negative Transform

```python
def apply_negative_transform(r):
    s = 255-r
    return s
```

## Task 3 - Applying each transformation to the loaded image

```python
log_trans_img = apply_log_transform(img, 1.15)
power_trans_img = apply_power_transform(img, 1.8, 127)
neg_trans_img = apply_negative_transform(img)
```

## Task 4 - Displaying all the generated images

In this case, matplotlib was used in order to display the images within the notebook. However, during testing, cv2.imshow() was also used, as displayed underneath the displayed images. Both approaches achieved the same results.

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 12), squeeze=False)

# Display the original image
axes[0, 0].imshow(img, cmap='gray')
axes[0, 0].set_title('Original Image')
axes[0, 0].axis('off')

# Display the logarithmic transformed image
axes[0, 1].imshow(log_trans_img, cmap='gray')
axes[0, 1].set_title('Log Image')
axes[0, 1].axis('off')

# Display the power transformed image
axes[1, 0].imshow(power_trans_img, cmap='gray')
axes[1, 0].set_title('Power Image')
axes[1, 0].axis('off')

# Display the negative tranformed image
axes[1, 1].imshow(neg_trans_img, cmap='gray')
axes[1, 1].set_title('Negative Image')
axes[1, 1].axis('off')
```

# Part 2: Area Processing

## Task 1 - Implementing a function which applies convolution to image depending on given kernel

```python
def apply_convolution(image, kernel):
    filt_img = cv2.filter2D(image, -1, kernel)
    return filt_img
```

## Task 2 - Designing different kernels with varying sizes

This is a list of dictionaries containing the kernels and their names, specifiying what is done to the image once filtering is applied.

```python
filters = [
    {'name': 'Edge Sharpening', 'kernel': np.array([[-1, -1, -1], 
                                                [-1, 8, -1], 
                                                [-1, -1, -1]])},
           
    {'name': 'Edge Detection 1', 'kernel': np.array([[-1, 0, 1], 
                                                    [0, 0, 0], 
                                                    [-1, 0, 1]])},

    {'name': 'Edge Detection 2', 'kernel': np.array([[0, 1, 0], 
                                                    [-1, 0, 1], 
                                                    [0, -1, 0]])},
    
    {'name': 'Box Blur (3x3)', 'kernel': np.ones((3,3))/9},
    {'name': 'Box Blur (5x5)', 'kernel': np.ones((5,5))/25},
    {'name': 'Box Blur (7x7)', 'kernel': np.ones((7,7))/49},
    {'name': 'Sharpen', 'kernel': np.array([[0, -1, 0],
                                           [-1, 5, -1],
                                           [0, -1, 0]])},

    {'name': 'Gaussian Blur (3x3)', 'kernel': np.array([[1, 2, 1],
                                                 [2, 4, 2],
                                                 [1, 2, 1]])/16},

    {'name': 'Gaussian Blur (5x5)', 'kernel': cv2.getGaussianKernel(ksize=5, sigma=1)},
    {'name': 'Gaussian Blur (7x7)', 'kernel': cv2.getGaussianKernel(ksize=7, sigma=1.5)}
]
```

## Task 3 + 4 - Applying convolution to the original image & displaying the generated images

Again, just like the previous task, matplotlib was used to display the images within the notebook. However, during testing, cv2.imshow() was also used. Both approaches achieved the same results.

```python
fig, axes = plt.subplots(nrows=4, ncols=3, figsize=(15, 15), squeeze=False)

# Display the original image
axes[0, 0].imshow(img, cmap='gray')
axes[0, 0].set_title('Original Image')
axes[0, 0].axis('off')

gs = axes[0,0].get_gridspec()
plt_size = gs.ncols * gs.nrows
filt_count = len(filters)

# Iterate through list of filters
for i, f in enumerate(filters):
    # Filter the image using the kernel being accessed
    filtered_img = apply_convolution(img, f['kernel'])

    # Display the filtered image
    row = (i+1) // gs.ncols
    col = (i+1) % gs.ncols
    axes[row, col].imshow(filtered_img, cmap='gray')
    axes[row, col].set_title(f['name'])
    axes[row, col].axis('off')

# Remove any blank subplots
for j in range(filt_count, plt_size-1):
    #print(j)
    row = (j+1) // gs.ncols
    col = (j+1) % gs.ncols
    #print(row, col)
    #fig.delaxes(axes[row][col])
    axes[row, col].set_visible(False)
```
